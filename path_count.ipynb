```python
import pandas as pd

df_startfinish= pd.read_csv('data/(신)누비자반납이력/12월_누비자반납이력.csv', sep=',', encoding='cp949')
df_startfinish.head()
```

# 2. 데이터 전처리 및 가공

```python
df_startfinish = df_startfinish.dropna(axis=0)
df_startfinish.head(10)
```

```python
df_startfinish = df_startfinish.astype({'new출발터미널' : 'int'})
df_startfinish = df_startfinish.astype({'new도착터미널' : 'int'})

df_startfinish.info()
```

```python
df_startfinish['출발~도착'] = df_startfinish['new출발터미널'].astype(str) + '~' + df_startfinish['new도착터미널'].astype(str)
df_startfinish
```

```python
df_startfinish.info()
```

```python
df_count = df_startfinish['출발~도착'].value_counts()
print(type(df_count))
```

```python
df_count = pd.DataFrame(df_count)
```

```python
df_count
```

```python
df_count.reset_index(inplace=True)
```

```python
df_count
```

```python
df_count.rename(columns={'index':'경로'}, inplace=True)
df_count.rename(columns={'출발~도착':'개수'}, inplace=True)
```

```python
df_count.info()
```

```python
df_count['new 출발'] = df_count.경로.str.extract(r'(\d+)')
df_count['new 도착'] = df_count.경로.str.extract(r'(~\d+)')
```

```python
df_count
```

```python
df_count['new 도착'] = df_count['new 도착'].str.replace(pat=r'[^\w]', repl=r'' , regex=True)
```

```python
df_count.head(30)
```

```python
df_count.info()
```

```python
print(df_count['new 출발'].isnull().sum())
print(df_count['new 도착'].isnull().sum())
```

```python
df_count = df_count.astype({'new 출발' : 'int'})
df_count = df_count.astype({'new 도착' : 'int'})
```

```python
df_count.info()
```

```python
df_count = df_count[['경로','new 출발','new 도착','개수']]
df_count
```

# 3. 가공 데이터 csv 저장

```python
df_count.to_csv('data/path/path_count12.csv', encoding='utf-8-sig',index=False)
```
