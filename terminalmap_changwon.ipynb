# 데이터 불러오기

```python
import pandas as pd

terminal_df = pd.read_csv('data/경상남도 창원시_누비자 터미널_20211231.csv', sep=',', encoding='cp949')
terminal_df.head()
terminal_df.info()
```

```python
bike_df = pd.read_csv('data/nubibike.csv', sep=',', encoding='cp949')
bike_df.head()
bike_df.info()
```

# 데이터 전처리

```python
merge_df = pd.merge(terminal_df, bike_df, how='right', left_on = '터미널명', right_on='터미널명')
merge_df.head()
```

```python
merge_df.info()
```

```python
merge_df.drop(columns=['번호_x', '구', '동', '주소_y', '보관대수_x'], inplace=True)
```

```python
merge_df = merge_df.rename(columns={'주소_x':'주소', '번호_y':'번호', '보관대수_y': '보관대수'})
merge_df.sort_values('번호')
```

# 터미널 위치 시각화

```python
import folium
```

```python
pd.Series(merge_df.columns)
```

```python
merge_df = merge_df.dropna(axis=0)
```

```python
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[35.2538, 128.6402],
    zoom_start=15
)

coords = merge_df[['번호','위도', '경도']]


marker_cluster = MarkerCluster().add_to(m)

for name, lat, long in zip(coords['번호'], coords['위도'], coords['경도']):
    folium.Marker([lat, long], popup= [name], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```
