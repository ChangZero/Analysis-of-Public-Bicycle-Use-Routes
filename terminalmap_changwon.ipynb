# 데이터 불러오기

```python
import pandas as pd

df_terminal = pd.read_csv('data/경상남도 창원시_누비자 터미널_20211231.csv', sep=',', encoding='cp949')
df_terminal.head()
df_terminal.info()
```

```python
df_crawling = pd.read_csv('data/nubibike.csv', sep=',', encoding='cp949')
df_crawling.head()
df_crawling.info()
```

# 데이터 전처리

```python
df_merge = pd.merge(df_terminal, df_crawling, how='right', left_on = '터미널명', right_on='터미널명')
df_merge.head()
```

```python
df_merge.info()
```

```python
df_merge.drop(columns=['번호_x', '구', '동', '주소_y', '보관대수_x'], inplace=True)
```

```python
df_merge = df_merge.rename(columns={'주소_x':'주소', '번호_y':'번호', '보관대수_y': '보관대수'})
df_merge.sort_values('번호')
```

# 터미널 위치 시각화

```python
import folium
```

```python
pd.Series(df_merge.columns)
```

```python
df_merge = df_merge.dropna(axis=0)
```

```python
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[35.2538, 128.6402],
    zoom_start=11
)

coords1 = df_merge[['번호','위도', '경도']]


marker_cluster1 = MarkerCluster().add_to(m)

for name, lat, long in zip(coords1['번호'], coords1['위도'], coords1['경도']):
    folium.Marker([lat, long], popup= [name], icon = folium.Icon(color="green")).add_to(marker_cluster1)
m
```

# 관광지 위치 시각화

```python
df_attraction = pd.read_csv('data/경상남도_창원시_주요관광지점 입장객_220627_워도.경도 ver.csv', sep=',', encoding='cp949')
df_attraction.head()
df_attraction.info()
```

```python
df_attraction = df_attraction.dropna(axis=0)
```

```python
coords2 = df_attraction[['관광지','위도', '경도']]


marker_cluster2 = MarkerCluster().add_to(m)

for name, lat, long in zip(coords2['관광지'], coords2['위도'], coords2['경도']):
    folium.Marker([lat, long], popup= [name], icon = folium.Icon(icon ="star", color="red")).add_to(marker_cluster2)
m
```
