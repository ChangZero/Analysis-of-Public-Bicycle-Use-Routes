```python
import pandas as pd

df_time= pd.read_csv('data/(신)누비자반납이력/3월_누비자반납이력.csv', sep=',', encoding='cp949')
df_time.head()
```

```python
df_time.info()
```

# 2. 데이터 추출 및 가공

```python
df_time['출발시간대'] = df_time.출발시간.str.split(':').str[0]
df_time['도착시간대'] = df_time.도착시간.str.split(':').str[0]
```

```python
df_time = df_time.dropna(axis=0)
df_time.head()
```

```python
df_time['출발시간대'].value_counts()
```

```python
df_time['도착시간대'].value_counts()
```

```python
df_time['출발전체시간'] = df_time['출발일'] + ' ' + df_time['출발시간']
df_time['도착전체시간'] = df_time['도착일'] + ' ' + df_time['도착시간']

df_time['이용시간'] = pd.to_datetime(df_time['도착전체시간']) - pd.to_datetime(df_time['출발전체시간'])
df_time = df_time.drop(['출발전체시간'], axis = 1)
df_time = df_time.drop(['도착전체시간'], axis = 1)

df_time.head()
```

```python
df_time.info()
df_time.describe()
```

```python
df_time = df_time.dropna(axis=0)

df_time = df_time.astype({'출발시간대' : 'int'})
df_time = df_time.astype({'도착시간대' : 'int'})

df_time = df_time.astype({'new출발터미널' : 'int'})
df_time = df_time.astype({'new도착터미널' : 'int'})

df_time.head()
df_time.describe()
```

```python
time_sum = df_time.groupby('new출발터미널')['이용시간'].sum()
```

```python
time_mean = df_time.groupby('new출발터미널')['이용시간'].mean()
```

```python
time_sum = pd.DataFrame(time_sum)
time_sum.rename(columns={'이용시간':'총이용시간'}, inplace=True)

time_mean = pd.DataFrame(time_mean)
time_mean.rename(columns={'이용시간':'평균이용시간'}, inplace=True)

time_sum.reset_index(inplace=True)
time_mean.reset_index(inplace=True)
time_mean = time_mean.drop(['new출발터미널'], axis = 1)

time_count = pd.concat([time_sum,time_mean],axis=1)

time_count
```

```python
df_use_time1 = df_time['출발시간대'].value_counts()
df_use_time1 = pd.DataFrame(df_use_time1)

df_use_time1 = df_use_time1.sort_index(ascending=False)
df_use_time1

df_use_time2 = df_time['도착시간대'].value_counts()
df_use_time2 = pd.DataFrame(df_use_time2)

df_use_time2 = df_use_time2.sort_index(ascending=False)
df_use_time2
```

```python
df_use_time = pd.concat([df_use_time1,df_use_time2],axis=1)
df_use_time
```

```python
df_use_time.reset_index(inplace=True)

df_use_time.rename(columns={'index':'이용시간대'}, inplace=True)
df_use_time.rename(columns={'출발시간대':'출발대수'}, inplace=True)
df_use_time.rename(columns={'도착시간대':'도착대수'}, inplace=True)

df_use_time
```

```python
df_use_time = df_use_time.astype({'이용시간대' : 'str'})
```

```python
time_count.info()
df_use_time.info()
```
